```python
import pandas as pd
import random
```

```python
leng=100000000
data={'a':(random.randint(0,100)for _ in range(leng)),
     'b':(random.randint(2,200)for _ in range(leng))}
```

```python
def hello_world():
    print('hello world')
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
df.std()
```

```python
import dask.dataframe as dd
```

```python
ddf=dd.from_pandas(df,npartitions=3)
```

```python
# it's empty
ddf.std()
```

```python
# to force an evaluation we use the compute method
ddf.std().compute()
```

```python
result=ddf.a.sum()-ddf.b.sum()
result
```

```python
result.compute()
```

```python
result.dask
```
